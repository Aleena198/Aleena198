# DATA ANALYSIS 

**What will be the  minimum and maximum sales for each store ?​**

```python
plt.figure(figsize = (18,6))
barplot = sns.barplot(x = 'Store',
           y = 'Weekly_Sales',
           data = wal_d,
           estimator = np.sum,
           ci = None,
           order = wal_d.groupby('Store').agg('sum').reset_index().sort_values(by = 'Weekly_Sales', ascending = True)['Store']).set_title('Sum of Weekly Sales of each Store')
plt.ylabel('WeeklySalesAggregate')
plt.show()
```



**What will be the customer's behavior on minimum and maximum sales stores during low/high temperature?​**

```python
store_min = wal_d[wal_d['Store'] == 33].sort_values(by = 'Date', ascending = True)
```

```python
store_min.head()
```

```python
store_max = wal_d[wal_d['Store'] == 20].sort_values(by = 'Date', ascending = True)
```

```python
store_max.describe()
```

```python
plt.figure(figsize = (15,8))
sns.lineplot(data=store_min,x='month',y='Weekly_Sales',hue='year',style='year',markers=True,palette='bright')
plt.title("WEEKLY SALES FOR THE MINIMUM SALES STORES RECORDED YEARS -2010 , 2011 , 2012")
plt.show()
```

```python
sns.regplot(x = store_min['Temperature'], y = store_min['Weekly_Sales'])
```

 **Which store has minimum/maximum sale during holiday? ​**

```python
min_sale = wal_d.loc[wal_d['Holiday_Flag'] == 1]['Weekly_Sales'].min()
wal_d.loc[wal_d['Weekly_Sales'] == min_sale]
```

```python
max_sale = wal_d.loc[wal_d['Holiday_Flag'] == 1]['Weekly_Sales'].max()
wal_d.loc[wal_d['Weekly_Sales'] == max_sale]
```
